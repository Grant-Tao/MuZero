```python
import gzip
import glob
import numpy as np

DATA_ITEM_LINES= 16+1+1+1


def unpack_plane(str_item) :
    x = np.zeros(361).astype(np.uint8)
    x[0:360] = np.unpackbits(np.frombuffer(bytearray.fromhex(str_item[0:90]), dtype=np.uint8))
    last_digit = str_item[90]
    if last_digit == "0" :
        x[360] = 0
    else :
        x[360] = 1
    return x


def unpack_unit(str_items) :
    x0 = unpack_plane(str_items[0])
    y0 = unpack_plane(str_items[8])
    stm = str_items[16][0]
    stm = np.uint8(stm)
    action = np.int16(361)
    probabilities = np.array(str_items[17].split()).astype(np.float32)
    winner = float(str_items[18])
    return (x0, y0, stm, action, probabilities, winner)

def process_chunkdata(chunkdata, games) :
    file_chunkdata = chunkdata.splitlines()
    game = []
    item = file_chunkdata[0:DATA_ITEM_LINES]
    str_items = [str(line, 'ascii') for line in item]

    x0l, y0l, stml, actionl, probsl, winnerl = unpack_unit(str_items)
        
    for i in range(DATA_ITEM_LINES, len(file_chunkdata), DATA_ITEM_LINES):
        item = file_chunkdata[i:i+DATA_ITEM_LINES]
        str_items = [str(line, 'ascii') for line in item]
        x0, y0, stm, action, probs, winner = unpack_unit(str_items)
        if max(x0)+max(y0)==0 :
            # begining of a game,  end of previous game
            game.append((x0l, y0l,stml, actionl, probsl, winnerl))
            games.append(game)
            game = []            
        else :
            actionl = np.argmax(y0-x0l)
            game.append((x0l, y0l,stml, actionl, probsl, winnerl))
        x0l, y0l,stml, actionl, probsl, winnerl = x0, y0,stm, action, probs, winner
        if i+DATA_ITEM_LINES==len(file_chunkdata) :
            game.append((x0l, y0l,stml, actionl, probsl, winnerl))
            games.append(game)
```

```python
data_prefix="data/"
chunks = glob.glob(data_prefix + "*.gz")

done = []
games = []

while len(chunks):
    filename = chunks.pop()
    with gzip.open(filename, 'rb') as chunk_file:
        done.append(filename)
        chunkdata = chunk_file.read()
        process_chunkdata(chunkdata, games)

print(len(games))
```
